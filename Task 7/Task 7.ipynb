```python
data = pd.read_csv('CO2_Emissions_metric_tons_per_capita_Canada_1960_to_2018.csv')
```

```python
data
```

```python
from sklearn.cluster import KMeans 
```

```python
kmeans = KMeans(3) 
```

```python
kmeans.fit(data) 
```

```python
clusters = kmeans.fit_predict(data) 
```

```python
clusters 
```

```python
clustered_data = data.copy() 
clustered_data['cluster_number'] = clusters 
clustered_data 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Year", fontsize=13) 
plt.ylabel("CO2 Emission (in metric tons per capita)", fontsize=13) 
plt.scatter(clustered_data['Year'], 
            clustered_data['CO2_Emissions'], 
            c=clustered_data['cluster_number'], 
            cmap='rainbow') 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
```

```python
from sklearn.cluster import KMeans 
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler() 
scaler.fit(data[['Year']]) 
data['Year'] = scaler.transform(data[['Year']]) 
 
scaler.fit(data[['CO2_Emissions']]) 
data['CO2_Emissions'] = scaler.transform(data[['CO2_Emissions']]) 
```

```python
data
```

```python
km = KMeans(n_clusters=3) 
y_predicted = km.fit_predict(data[['Year','CO2_Emissions']]) 
y_predicted 
```

```python
data['cluster_number'] = y_predicted 
data
```

```python
km.cluster_centers_ 
```

```python
df1 = data[data.cluster_number==0] 
df2 = data[data.cluster_number==1] 
df3 = data[data.cluster_number==2] 
 
plt.scatter(df1.CO2_Emissions, df1['Year'], color='Grey') 
plt.scatter(df2.CO2_Emissions, df2['Year'], color='Orange') 
plt.scatter(df3.CO2_Emissions, df3['Year'], color='Green') 
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='Blue', marker= '*', label='centroid') 
 
plt.legend
```
